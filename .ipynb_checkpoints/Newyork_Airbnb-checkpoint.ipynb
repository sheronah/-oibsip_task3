```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\Sheronah baBy\datasets\AB_NYC_2019.csv")
df
```

```python
print(df.info())
```

```python
print(df.head())
```

```python
print(df.describe())
```

```python
missing_values = df.isnull().sum()
missing_values
```

```python
df['reviews_per_month'] = df['reviews_per_month'].fillna(0)
df['name'] = df['name'].fillna('Unknown')
df['host_name'] = df['host_name'].fillna('Unknown')
df['last_review'] = df['last_review'].fillna(pd.NaT)
df
```

```python
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
```

```python
# Convert last_review to datetime
df['last_review'] = pd.to_datetime(df['last_review'])
df
```

```python
# Examine price distribution
print(df['price'].describe())
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x=df['price'])
plt.title('Price Distribution Before Cleaning')
plt.show()

```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['price'], bins=100)
plt.xlim(0, 1000)  # Limit to $1000 to see the main distribution
plt.title('Price Histogram (Up to $1000)')
plt.show()
```

```python
# Remove listings with price = 0 (likely errors)
df = df[df['price'] > 0]
```

```python
# Calculate percentiles
price_95 = df['price'].quantile(0.95)  # ~$350
price_99 = df['price'].quantile(0.99)
```

```python
df.loc[:, 'price'] = df['price'].clip(upper=price_99)
```

```python
print(df['price'].describe())
```

```python
df.loc[:,'availability_365'] = df['availability_365'].clip(0, 365)
df
```

```python
# Minimum nights should be at least 1
df.loc[:,'minimum_nights'] = df['minimum_nights'].clip(1)
df
```

```python
# Days since last review
df.loc[:,'days_since_last_review'] = (pd.to_datetime('2019-07-08') - df['last_review']).dt.days
df
```

```python
 # Categorize price ranges
df.loc[:,'price_category'] = pd.cut(df['price'], 
    bins=[0, 100, 200, 300, float('inf')],
    labels=['Budget', 'Moderate', 'Expensive', 'Luxury'])
df
```

```python
# Verify no missing values remain
print(df.isnull().sum())

# Check data types
print(df.dtypes)

# Verify no duplicates
print(df.duplicated().sum())

# Check descriptive statistics
print(df.describe())
```
