```python
# Check missing values percentage
missing_percent = df.isna().sum() / len(df) * 100
print("Missing values percentage:\n", missing_percent)
```

```python
# Handle missing values
df.fillna({'description': 'No description'}, inplace=True)
```

```python
# Clean title and channel names
text_cols = ['title', 'channel_title', 'description', 'tags']
for col in text_cols:
    df[col] = df[col].str.strip()  # Remove whitespace
    df[col] = df[col].str.replace(r'\s+', ' ', regex=True)  # Replace multiple spaces
```

```python
# Clean tags (convert to list and standardize format)
df['tags'] = df['tags'].str.replace('"', '').str.split('|')
```

```python
# Extract useful features from publish_time
df['publish_time'] = pd.to_datetime(df['publish_time'])
df['publish_year'] = df['publish_time'].dt.year
df['publish_month'] = df['publish_time'].dt.month
df['publish_day'] = df['publish_time'].dt.day
df['publish_hour'] = df['publish_time'].dt.hour
```

```python
# Check for duplicate videos
print("Duplicate video_ids:", df.duplicated(subset=['video_id']).sum())
```

```python
# Remove exact duplicates
df.drop_duplicates(subset=['video_id'], keep='first', inplace=True)
```

```python
# Check for near-duplicates (same title and channel)
near_dups = df.duplicated(subset=['title', 'channel_title'], keep=False)
if near_dups.sum() > 0:
    print(f"Found {near_dups.sum()} potential near-duplicates")
    # Option: Keep the most viewed version
    df.sort_values('views', ascending=False, inplace=True)
    df.drop_duplicates(subset=['title', 'channel_title'], keep='first', inplace=True)
```

```python
# Convert boolean-like columns to proper booleans
bool_cols = ['comments_disabled', 'ratings_disabled', 'video_error_or_removed']
for col in bool_cols:
    df[col] = df[col].astype(bool)
```

```python
# Standardize numerical columns (remove any commas, convert to numeric)
num_cols = ['views', 'likes', 'dislikes', 'comment_count']
for col in num_cols:
    if df[col].dtype == 'object':
        df[col] = df[col].str.replace(',', '').astype(float)
```

```python
# Visualize numerical columns for outliers
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[num_cols])
plt.yscale('log')  # Using log scale due to large ranges
plt.title("Boxplot of Numerical Columns (log scale)")
plt.xticks(rotation=45)
plt.show()
```

```python
# Function to cap outliers
def cap_outliers(series, lower_quantile=0.01, upper_quantile=0.99):
    lower = series.quantile(lower_quantile)
    upper = series.quantile(upper_quantile)
    return series.clip(lower, upper)

# Cap outliers in engagement metrics
for col in num_cols:
    df[col] = cap_outliers(df[col])
```

```python
# Create engagement metrics
df['like_ratio'] = df['likes'] / (df['likes'] + df['dislikes']).replace(0, 1)
df['comment_ratio'] = df['comment_count'] / df['views'].replace(0, 1)
df['engagement_rate'] = (df['likes'] + df['dislikes'] + df['comment_count']) / df['views'].replace(0, 1)
```

```python
df['publish_time'] = pd.to_datetime(df['publish_time']) 
today = pd.Timestamp.now(tz='UTC')  
df['video_age_days'] = (today - df['publish_time']).dt.days
```

```python
df['tag_count'] = df['tags'].apply(len)
```
